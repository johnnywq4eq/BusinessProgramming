## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### 💡 Pro Tip: Be specific about requirements!

```python
"""
My Budget Tracker Prompt:
Write a Python program that:
1) Asks the user for monthly income
2) Allows the user to enter up to 3 expenses
3) Calculates remaining budget (income - total expenses)
4) Prints income, each expense, total expenses, and remaining budget
5) Shows a warning message if remaining budget is negative
Keep the program simple and beginner-friendly.
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# [Paste your code here]name = input("What is your name? ")
income = float(input("Enter your monthly income (AUD): "))

expense1 = float(input("Enter expense 1 amount (AUD): "))
expense2 = float(input("Enter expense 2 amount (AUD): "))
expense3 = float(input("Enter expense 3 amount (AUD): "))

total_expenses = expense1 + expense2 + expense3
remaining = income - total_expenses

print("\n" + "="*40)
print(f"Budget Summary for {name}")
print("="*40)
print(f"Income: ${income:.2f}")
print(f"Expense 1: ${expense1:.2f}")
print(f"Expense 2: ${expense2:.2f}")
print(f"Expense 3: ${expense3:.2f}")
print(f"Total Expenses: ${total_expenses:.2f}")
print(f"Remaining Budget: ${remaining:.2f}")

if remaining < 0:
    print("⚠️ WARNING: Your budget is negative! You spent more than your income.")
else:
    print("✅ Good job! You are within your budget.")
```
